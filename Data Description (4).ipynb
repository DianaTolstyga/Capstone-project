After we get the venues list for one neighbourhood, we will put the data including venue's name, longitude/latitude, and category type into a dataframe:

```python
def get_category_type(row):
    try:
        categories_list=row['categories']
    except:
        categories_list=row['venue.categories']
        
    if len(categories_list)==0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues=results['response']['groups'][0]['items']
nearby_venues=json_normalize(venues)
filtered_columns=['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues=nearby_venues.loc[:, filtered_columns]

nearby_venues['venue.categories']=nearby_venues.apply(get_category_type, axis=1)
nearby_venues.columns=[col.split(".")[-1] for col in nearby_venues.columns]
nearby_venues.head(5)
```

```python
print("Number of venues near {}: {}".format(Paris_data.loc[0, 'Neighbourhood'], nearby_venues.shape[0]))
```

Now, we will repeat the process for all the selected neighbourhoods in Paris and create a dataframe "Paris_venues":

```python
def getNearbyVenues(names, latitudes, longitudes, radious=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
    
        url='https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={}, {}&radius={}&limit={}'.format(
             CLIENT_ID, CLIENT_SECRET, VERSION, lat, lng, radius, LIMIT)
    
        results=requests.get(url).json()["response"]['groups'][0]['items']
    
        venues_list.append([(
             name, 
             lat, 
             lng, 
             v['venue']['name'], 
             v['venue']['location']['lat'],
             v['venue']['location']['lng'], 
             v['venue']['categories'][0]['name']) for v in results ])
    
    nearby_venues=pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns=['Neighborhood','Neighborhood Latitude', 'Neighborhood Longitude', 'Venue', 'Venue Latitude', 'Venue Longitude', 'Venue Category']
    
    return(nearby_venues)
        
```

```python
Paris_venues=getNearbyVenues(names=Paris_data['Neighbourhood'], latitudes=Paris_data['Latitude'], longitudes=Paris_data['Longitude'], radious=500)

Paris_venues.head()
```

```python
Paris_venues.shape
```

We have got around 460 venues in total which we will further invastigate. Let us group each venue category by neighbourhood and aggregate the values by taking a mean across the column:

```python
Paris_twohot=pd.get_dummies(Paris_venues[['Neighborhood']], prefix="", prefix_sep="")
Paris_twohot['Venue Category']=Paris_venues['Venue Category']
Paris_twohot['Venue Latitude']=Paris_venues['Venue Latitude']
Paris_twohot['Venue Longitude']=Paris_venues['Venue Longitude']
Paris_twohot.rename({'2nd arrondissement of Paris': '1nd Arrondissement of Paris'}, axis=1, inplace=True)
Paris_twohot
```

```python
Paris_grouped_bn=Paris_twohot.groupby('Venue Category').mean().reset_index()
Paris_grouped_bn
```

Now we have all the data required to proceed k-clustering analysis to classify the venues from each neighbourhood. However, before we complete the analysis, let us first choose the best k using elbow method. It should be an argument for which the distortion function is dicreasing significantly:

```python
from sklearn.cluster import KMeans
Paris_grouped_bn_clustering=Paris_grouped_bn.iloc[:,1:6]
distortions=[]
K=range(1,10)
for k in K:
    KmeansModel=KMeans(n_clusters=k, random_state=0).fit(Paris_grouped_bn_clustering)
    distortions.append(KmeansModel.inertia_)
    
distortions
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,6))
plt.plot (K, distortions, 'bx-')
plt.xlabel('K')
plt.ylabel('Distortions')
plt.show()
```

Thus for the given data, we conclude that the optimal number of clusters is 5. Let us complete the k-clustering analysis:

```python
kclusters=5
Paris_grouped_bn_clustering=Paris_grouped_bn.iloc[:,1:6]
kmeans=KMeans(n_clusters=kclusters, random_state=0).fit(Paris_grouped_bn_clustering)
kmeans.labels_[0:10]
```

```python
Paris_grouped_bn.insert(0, 'Cluster Labels', kmeans.labels_)
Paris_grouped_bn.head()
```

Now as we have all clusters defined for each venue category, let us visualize the results:

We can now choose the best neighborhood in Paris. To do that we will use a combined rating:
1. The first rating is using a cluster analysis. We have 5 clusters of venue categories and our preference is to live at the center of the city at the most diverse community which means that we want venues from different clusters to present in our neighbourhood. That is why the metrics is taken into consideration the variaty of clusters in one neigbourhood. The neighbourhood with venues from all 5 clusters will get the highest rating.
2. The second rating is using our personal ranking system. We simply take all venue categories from the venues list for all three cities and manually rank each category from 1 to 10 based on our personal preferences i.e. science museums and cultural places are getting the highest score because we like science, while smoke shops liquor storesare getting score 1 because we are not smoking and drinking fans.

To get the first rating we make some transformations with Paris data and then rate each neighbourhood based on the following assumptions:
(1) if the value for a venue in a neighbourhood column is less than 1.0, we assume that this type of venue is underrepresented in a neignbourhood and it earns zero point; 
(2) if the value for a venue in a neighbourhood column is more or equal to 1.0. we assume that this type of venue is presented in a neighbourhood and we keep it and its cluster label. And for each such cluster, a neighbourhood earns 10 points, i.e. if the neignbourhood has venues from all 5 clusters, it will get 50 points:

```python
Paris_grouped_bn_grouped=Paris_grouped_bn.iloc[:,0:7].groupby('Cluster Labels').sum().reset_index()
Paris_grouped_bn_grouped
```

```python
Paris_cluster_rating=np.asarray([Paris_grouped_bn_grouped.columns[1:6].tolist(),[0,0,0,0,0]])
number=np.asarray(Paris_grouped_bn_grouped[Paris_grouped_bn_grouped > 1.0].count())
i=1
for i in range(5):
       Paris_cluster_rating[1][i-1]=number[i]*10
Paris_cluster_rating
```

```python
for line in Paris_cluster_rating:
    print('\t'.join(map(str,line)))
```

```python
print('The neighbourhood {} has the highest score which is {} '.format(Paris_cluster_rating[0][np.argmax(Paris_cluster_rating[1])],max(Paris_cluster_rating[1])))
```

We see that Quartier latin has the highest rating while Eiffel Tower has the lowest one.


Now we will get the second rating based on our personal preferences. To do that we first need to get the list of all venues and manually rank each category from 1 to 10. The combined list of categories for all the cities has 206 different categories which we manually rate; its sum for each neighbourhood will lead to a total rank.

```python
#Paris_grouped_bn.insert(0, 'Cluster Labels', kmeans.labels_) #use this line if Cluster Labels data are not added to a table
Paris_grouped_bn.head()
```

Here is the ranking list of all venues categories:

We will combine this list with Paris data and then define the ranking function:

```python
Paris_merged=venues_categories
Paris_merged=Paris_merged.join(Paris_grouped_bn.set_index('Venue Category'), on='Venue Category')
Paris_merged.dropna(axis=0, inplace=True)
Paris_merged.head()
```

Function to define a rank of each neighbourhood is a product of Ranking column and the data for each neighbourhood:

```python
Paris_rank=[]
i=1
l=Paris_data.shape[0]
for i in range(l):
    rank=Paris_merged.iloc[:,4+i]*Paris_merged['Ranking']
    Paris_rank.append(rank.sum()),
Paris_rank
```

Now we have the second rating calculated for each neighbourhood. Let us now sum up these two ratings to get the total Paris rating:

```python
total_Paris_rank=[]
for (item1, item2) in zip(Paris_cluster_rating[1].tolist(), Paris_rank):
    total_Paris_rank.append(int(item1)+item2)

total_Paris_rank
```

```python
print("The neighbourhood ' {} ' has the highest rank and its score is {}".format(Paris_cluster_rating[0][np.argmax(total_Paris_rank)], max(total_Paris_rank)))
```
